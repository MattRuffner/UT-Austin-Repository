```python
X_scaled = StandardScaler()
X_train = X_scaled.fit_transform(X_train)
X_test = X_scaled.transform(X_test)
```

```python
## 1A
# training set RMSE
mlp_model5 = MLPRegressor(activation = 'tanh', solver = 'sgd', learning_rate='constant', random_state=42, batch_size= 20, learning_rate_init = 0.005, hidden_layer_sizes =(5,)).fit(X_train, y_train)  
y_pred5a = mlp_model5.predict(X_train)
print("training set RMSE for 5 hidden layers:", np.sqrt(mean_squared_error(y_train, y_pred5a)))

# test set RMSE
mlp_model5 = MLPRegressor(activation = 'tanh', solver = 'sgd', learning_rate='constant', random_state=42, batch_size= 20, learning_rate_init = 0.005, hidden_layer_sizes =(5,)).fit(X_train, y_train)  
y_pred5e = mlp_model5.predict(X_test)
print("test set RMSE for 5 hidden layers:", np.sqrt(mean_squared_error(y_test, y_pred5e)))
```

```python
## 1B
# training set RMSE
mlp_model15 = MLPRegressor(activation = 'tanh', solver = 'sgd', learning_rate='constant', random_state=42, batch_size= 20, learning_rate_init = 0.005, hidden_layer_sizes =(15,)).fit(X_train, y_train)  
y_pred15a = mlp_model15.predict(X_train)
print("training set RMSE for 15 hidden layers:", np.sqrt(mean_squared_error(y_train, y_pred15a)))

# test set RMSE
mlp_model15 = MLPRegressor(activation = 'tanh', solver = 'sgd', learning_rate='constant', random_state=42, batch_size= 20, learning_rate_init = 0.005, hidden_layer_sizes =(15,)).fit(X_train, y_train)  
y_pred15e = mlp_model15.predict(X_test)
print("test set RMSE for 15 hidden layers:", np.sqrt(mean_squared_error(y_test, y_pred15e)))
```

```python
## 1C
# training set RMSE
mlp_model50 = MLPRegressor(activation = 'tanh', solver = 'sgd', learning_rate='constant', random_state=42, batch_size= 20, learning_rate_init = 0.005, hidden_layer_sizes =(50,)).fit(X_train, y_train)  
y_pred50a = mlp_model50.predict(X_train)
print("training set RMSE for 50 hidden layers:", np.sqrt(mean_squared_error(y_train, y_pred50a)))

# test set RMSE
mlp_model50 = MLPRegressor(activation = 'tanh', solver = 'sgd', learning_rate='constant', random_state=42, batch_size= 20, learning_rate_init = 0.005, hidden_layer_sizes =(50,)).fit(X_train, y_train)  
y_pred50e = mlp_model50.predict(X_test)
print("test set RMSE for 50 hidden layers:", np.sqrt(mean_squared_error(y_test, y_pred50e)))
```

```python
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
median_train = imp_median.fit_transform(X_train)
median_test = imp_median.fit_transform(X_test)

lin_reg_median = LinearRegression().fit(median_train, y_train)
y_pred_median = lin_reg_median.predict(median_test)
print("R^2 score with medians imputed for NaNs:",r2_score(y_test, y_pred_median))
```

```python
imp_most_freq = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
most_freq_train = imp_most_freq.fit_transform(X_train)
most_freq_test = imp_most_freq.fit_transform(X_test)

lin_reg_freq = LinearRegression().fit(most_freq_train, y_train)
y_pred_freq = lin_reg_freq.predict(most_freq_test)
print("R^2 score with most frequent value imputed for NaNs:", r2_score(y_test, y_pred_freq))
```

```python
imp_knn2 = KNNImputer(n_neighbors=2)
knn2_train = imp_knn2.fit_transform(X_train)
knn2_test = imp_knn2.fit_transform(X_test)

lin_reg_knn2 = LinearRegression().fit(knn2_train, y_train)
y_pred_knn2 = lin_reg_knn2.predict(knn2_test)
print("R^2 score with 2 nearest neighbor value imputed for NaNs:",r2_score(y_test, y_pred_knn2))
```

```python
imp_knn5 = KNNImputer(n_neighbors=5)
knn5_train = imp_knn5.fit_transform(X_train)
knn5_test = imp_knn5.fit_transform(X_test)

lin_reg_knn5 = LinearRegression().fit(knn5_train, y_train)
y_pred_knn5 = lin_reg_knn5.predict(knn5_test)
print("R^2 score with 5 nearest neighbor value imputed for NaNs:",r2_score(y_test, y_pred_knn5))
```

```python
imp_knn10 = KNNImputer(n_neighbors=10)
knn10_train = imp_knn10.fit_transform(X_train)
knn10_test = imp_knn10.fit_transform(X_test)

lin_reg_knn10 = LinearRegression().fit(knn10_train, y_train)
y_pred_knn10 = lin_reg_knn10.predict(knn10_test)
print("R^2 score with 10 nearest neighbor value imputed for NaNs:",r2_score(y_test, y_pred_knn10))
```

```python
imp_knn20 = KNNImputer(n_neighbors=20)
knn20_train = imp_knn20.fit_transform(X_train)
knn20_test = imp_knn20.fit_transform(X_test)

lin_reg_knn20 = LinearRegression().fit(knn20_train, y_train)
y_pred_knn20 = lin_reg_knn20.predict(knn20_test)
print("R^2 score with 20 nearest neighbor value imputed for NaNs:",r2_score(y_test, y_pred_knn20))
```

# Conclusion
Imputing our NaN values with the average of the 20 nearest known neighbor values gives us an R^2 value of .6378. Simple imputer gives us slightly lower R^2 than KNN imputer since the simple imputer is looking at the dataset as a whole leading to overgeneralization while KNN imputer takes looks at a more local approximation of the value in question leading to a more representative value of the missing data.
